**Q2. Phân khúc sản phẩm (segment) nào trong products.csv có tỷ suất lợi nhuận gộp trung bình cao nhất, với công thức (price − cogs)/price?**
A) Premium
B) Performance
C) Activewear
D) Standard

```python
df_product.head()
```

```python
df_product.info()
print(df_product.isnull().sum())
print(df_product.duplicated().sum())
print(df_product.nunique().sum())
```

```python
#công thức đề cho(price - cogs) / price
df_product['gross_margin'] = (df_product['price'] - df_product['cogs']) / df_product['price']
#Tỷ suất lợi nhuận gộp trung bình
segment= df_product.groupby('segment')['gross_margin'].mean().sort_values(ascending=False)
print('Tỷ suất lợi nhuận gộp trung bình theo từng segment',segment)
#Tìm ra cao nhất
max_segment=segment.index[0]
max_value=segment.iloc[0]
print(f'Phân khúc {max_segment} có tỷ suất lợi nhuận cao nhất với {max_value}')
```

Q7. Vùng (region) nào trong geography.csv tạo ra tổng doanh thu cao nhất trong sales_train.csv?
A) West
B) Central
C) East
D) Cả ba vùng có doanh thu xấp xỉ bằng nhau

```python
#công thức revenue=(quantity*price)-discount
df_items['revenue']=(df_items['quantity']*df_items['unit_price'])-df_items['discount_amount']
print(df_items['revenue'])
```

```python
#items gộp với order qua order_id, gộp với geo qua zip
df_merge = pd.merge(df_items, df_or, on='order_id', how='inner')
df_merge = pd.merge(df_merge, df_geo, on='zip', how='inner')
#revenue theo từng region
revenue_by_region = df_merge.groupby('region')['revenue'].sum().sort_values(ascending=False)
print('Tổng doanh thu của từng vùng:',revenue_by_region)
max_region = revenue_by_region.index[0]
max_revenue = revenue_by_region.iloc[0]
print(f'Vùng {max_region} tạo ra doanh thu cao nhất với tổng cộng {max_revenue:,.2f}')
```
